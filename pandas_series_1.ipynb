```python
import numpy as np
import pandas as pd
```

```python
## series from list

country = ['India','Pakistan','USA','Nepal','Srilanka']
pd.Series(country)
```

```python
run = [13,24,56,78,100]
pd.Series(run)
```

```python
# custom index

marks = [67,57,89,100]
subjects = ['maths','english','science','hindi']

pd.Series(marks, index = subjects)
```

```python
# setting a name
pd.Series(marks, index = subjects,name = "student's marks")
```

```python
##series from dictionary

marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}
pd.Series(marks)
```
